```python
# Pandas,NumPy, Matplotlib and Seaborn are imported ready for analysis:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# The data for March 2022 is imported and then sense-checked:

mar22 = pd.read_csv('2022-03-metropolitan-street.csv')

mar22.head()
```

```python
# The same is now done for the rest of the year (2022 to 2023) - each month was provided in a separate csv,
# so needs to be imported separately:

apr22 = pd.read_csv('2022-04-metropolitan-street.csv')
may22 = pd.read_csv('2022-05-metropolitan-street.csv')
jun22 = pd.read_csv('2022-06-metropolitan-street.csv')
jul22 = pd.read_csv('2022-07-metropolitan-street.csv')
aug22 = pd.read_csv('2022-08-metropolitan-street.csv')
sep22 = pd.read_csv('2022-09-metropolitan-street.csv')
oct22 = pd.read_csv('2022-10-metropolitan-street.csv')
nov22 = pd.read_csv('2022-11-metropolitan-street.csv')
dec22 = pd.read_csv('2022-12-metropolitan-street.csv')
jan23 = pd.read_csv('2023-01-metropolitan-street.csv')
feb23 = pd.read_csv('2023-02-metropolitan-street.csv')
```

```python
# All the DataFrames are concatenated by row into one single DataFrame called all_data_met:

all_data_met = pd.concat([mar22, apr22, may22, jun22, jul22, aug22, sep22, oct22, nov22, dec22, jan23, feb23], axis=0)
```

```python
# The same is done with data from the City of London police, to obtain a full London dataset.
# No data is available for June 2022 and this therefore limits accuracy of any conclusions:

city_mar22 = pd.read_csv('2022-03-city-of-london-street.csv')
city_apr22 = pd.read_csv('2022-04-city-of-london-street.csv')
city_may22 = pd.read_csv('2022-05-city-of-london-street.csv')
city_jul22 = pd.read_csv('2022-07-city-of-london-street.csv')
city_aug22 = pd.read_csv('2022-08-city-of-london-street.csv')
city_sep22 = pd.read_csv('2022-09-city-of-london-street.csv')
city_oct22 = pd.read_csv('2022-10-city-of-london-street.csv')
city_nov22 = pd.read_csv('2022-11-city-of-london-street.csv')
city_dec22 = pd.read_csv('2022-12-city-of-london-street.csv')
city_jan23 = pd.read_csv('2023-01-city-of-london-street.csv')
city_feb23 = pd.read_csv('2023-02-city-of-london-street.csv')
```

```python
# All the DataFrames are concatenated by row into one single DataFrame called all_data_city:

all_data_city = pd.concat([mar22, apr22, may22, jul22, aug22, sep22, oct22, nov22, dec22, jan23, feb23], axis=0)
```

```python
# The Met and City of London data is concatenated into a single dataset:

all_data = pd.concat([all_data_met, all_data_city], axis=0)
```

```python
# The data is sense checked and we can see there are 2,131,650 crimes for the year:

all_data.info()
```

```python
all_data.head()
```

```python
# The data is checked for null values.  There are numerous null values, but these will be left as NaN as they do not
# interfere with the analysis.  However, it shows that the dataset is not complete or 100% reliable.  But, it seems
# we can rely on the total number of bicycle thefts as this contains no null values.  We just have to use caution
# where top boroughs etc are concerned, as there are a large number of crimes with no location information.

all_data.isnull().sum()
```

```python
# A check is donw to see which LSOAs (Local Super Output Areas) are top:

all_data['LSOA name'].value_counts()
```

```python
# The code needs strip out the LSOA name so that the borough names (e.g. Westminster) align with our
# London cycle survey dataset.
# The last 5 characters (4 code characters and space) are removed from the values in the LSOA name column:

borough = pd.DataFrame(all_data['LSOA name'].str[:-5])
borough.rename(columns={'LSOA name': 'borough'}, inplace=True)

borough.head()
```

```python
# The clean column is appended to the original dataset:

all_data_clean = pd.concat([all_data, borough], axis=1)

all_data_clean.head()
```

```python
# The full data set is exported to Excel to save the full data set for future use or in other programmes such as R
# or Tableau.

all_data_clean.to_csv("all_data.csv")
```

```python
# Offence types in the Metropolitan Police/City of London area are counted - violence and sexual offences are top. 
# There were 34,347 bicycle thefts reported in the 12 month period.

top_offences = pd.DataFrame(all_data_clean['Crime type'].value_counts())
top_offences.index.names = ['Crime Type']
top_offences.rename(columns={'Crime type': 'Count'}, inplace=True)

top_offences

top_offences.to_csv("count_of_top_london_crime_types.csv") # Export to csv for any later use.
```

```python
# The top locations for crime are counted.  Westminster comes a clear top:

top_loc = all_data_clean['borough'].value_counts()

top_loc.head(30)
```

```python
# The data is subset to just examine bicycle theft:

bicycle_theft = all_data_clean[all_data_clean['Crime type'] == 'Bicycle theft']

bicycle_theft.head()
```

```python
# The bicycle theft data is exported to csv for use with Tableau to map the offence locations:

bicycle_theft.to_csv("bicycle_theft_data.csv")
```

```python
# The bike theft data is checked for null values.
# No crime IDs are null, which is useful for mapping the offences with Tableau.
# Those without location data, however, cannot be mapped.

bicycle_theft.isnull().sum()
```

```python
# Further subsetting this in order to count how many per borough shows that Hackney is top.
# A new data frame containing only the top 10 is created:

top_bike_locs = bicycle_theft.groupby(['borough'])['Crime type'].agg('count'
                                                                    ).sort_values(ascending=False).reset_index()

top_10_bike_boroughs = top_bike_locs.head(10)

top_10_bike_boroughs
```

```python
# A barplot is generate to visualise the top 10 boroughs for bike theft:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(15, 15)})
ax.set_title("Top Locations in London for Bicycle Theft (Mar 22 - Feb 23)")
ax.set_xlabel("London Borough")
ax.set_ylabel("Total Thefts")
plt.xticks(rotation=90)

sns.barplot(x='borough', y='Crime type', data=top_10_bike_boroughs)

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')

fig.savefig('top_bike_theft_borough_12months.png')
```

```python
# The bicycle data is further filtered to show month of offence by borough as well:

bike_borough_month = bicycle_theft.groupby(['borough', 'Month'
                                           ])['Crime type'].agg('count').reset_index()

bike_borough_month
```

```python
# A line chart is created to see the change in numbers over the data period:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(12, 15)})
ax.set_title("Total Bicycle Thefts by Month (March 22 - Feb 23)")
ax.set_xlabel("Month")
ax.set_ylabel("Total Bicycle Thefts")
plt.xticks(rotation=90) # x-axis labels are rotated 90 degrees to be more readble.

sns.lineplot(x='Month', y='Crime type', data=bike_borough_month, ci=None)

fig.savefig('all_bike_thefts_by_month.png')
```

```python
# The data is then further filtered to only contain the top 10 boroughs for bicycle theft:

boroughnames = ['Westminster', 'Hackney', 'Southwark', 'Tower Hamlets', 'Camden', 'Lambeth', 'Islington', 'Hammersmith and Fulham', 'Wandsworth', 'Richmond upon Thames']
bike_top_boroughs_month = bike_borough_month[bike_borough_month['borough'].isin(boroughnames)]
```

```python
bike_top_boroughs_month
```

```python
# This is used to create a line chart showing each individual top 10 borough by month.
# There is a dip in June 2022 - there was no data provided for this month for the City of London Police, so 
# it is an anomaly.
# However, it appears that thefts dropped in December - perhaps due to weather and less cycling going on?

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(12, 15)})
ax.set_title("Total Bicycle Thefts in Top 10 Boroughs by Month (March 22 - Feb 23)")
ax.set_xlabel("Month")
ax.set_ylabel("Total Bicycle Thefts")
plt.xticks(rotation=90) # x-axis labels are rotated 90 degrees to be more readble.

sns.lineplot(x='Month', y='Crime type', hue='borough', data=bike_top_boroughs_month, ci=None)

fig.savefig('top10_borough_bike_thefts_by_month.png')
```
