### Let's start by loading the dataset and conducting an exploratory data analysis (EDA).

```python
import pandas as pd
df = pd.read_csv('central_london_clean.csv')
df.head()
```

```python
df.describe(include='all')
```

### Now, let's visualize the data to get a better understanding of the distribution and relationships between different variables. We'll start by looking at the trend of cycling over the years.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='year', y='total_cycles')
plt.title('Trend of Cycling Over the Years')
plt.show()
```

### Based on the trend analysis, we observed that the number of cyclists in London has been relatively stable over the years. There doesn't appear to be a significant upward or downward trend in the total number of cyclists. This suggests that while cycling is a popular mode of transport in London, there hasn't been a significant increase in its popularity over the years covered by the dataset.

### Next, let's perform a factor analysis to understand which factors might be influencing the number of cyclists. We'll start by examining the impact of weather on cycling.

```python
plt.figure(figsize=(35, 20))
sns.boxplot(data=df, x='Weather', y='total_cycles')
plt.title('Impact of Weather on Cycling')
plt.xticks(rotation=90)
plt.show()
```

### As we can see, the weather has a significant impact on cycling. The number of cyclists tends to be higher on dry days compared to wet or snowy days.



### Next, let's examine the impact of the day of the week on cycling.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Day', y='total_cycles', order=['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun'])
plt.title('Impact of Day of the Week on Cycling')
plt.show()
```

### It appears that the number of cyclists is relatively consistent across weekdays, with a slight decrease on the weekends. This could be due to commuting patterns during the workweek.



### Here's the impact of the time of day on cycling:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Period', y='total_cycles')
plt.title('Impact of Time of Day on Cycling')
plt.xticks(rotation=90)
plt.show()
```

### From the plot, we can see that the number of cyclists varies significantly depending on the time of day. The highest number of cyclists are observed during the morning peak (07:00-10:00) and the afternoon peak (16:00-19:00), which aligns with typical commuting hours. The number of cyclists is relatively lower during the inter-peak period (10:00-16:00) and significantly lower during the night (19:00-07:00).

### This suggests that a significant portion of cycling in London is likely for commuting purposes.
