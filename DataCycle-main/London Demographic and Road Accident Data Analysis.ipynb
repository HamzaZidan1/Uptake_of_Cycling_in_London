```python
# Pandas,NumPy, Matplotlib and Seaborn are imported ready for analysis:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# The data is imported (first of two spreadsheets) and sense-checked:

data1 = pd.read_excel('demographic_data_for_analysis1.xls')

data1.head()
```

```python
data1.describe()
```

```python
data1.info()
```

```python
# The data is checked for null values.
#  There are no null values.

data1.isnull().sum()
```

```python
# The names column is cleaned to show only borough name (rather than Local Super Output Area):

borough1 = pd.DataFrame(data1['Names'].str[:-5])
borough1.rename(columns={'Names': 'Borough'}, inplace=True)

borough1.head()
```

```python
# The clean column is appended to the original dataset:

data1_clean = pd.concat([data1, borough1], axis=1)

data1_clean.head()
```

```python
# The same process is then followed with the second spreadsheet of data:

data2 = pd.read_excel('demographic_data_for_analysis2.xlsx')

data2.head()
```

```python
data2.describe()
```

```python
data2.info()
```

```python
data2.isnull().sum()
```

```python
# The 'Names' column is deleted from this dataset, as this data frame will be joined with the first:

data2.drop('Names', axis=1, inplace=True)

data2.head()
```

```python
# The two data frames are concatenated together:

london_demo_data = pd.concat([data1_clean, data2], axis=1)

london_demo_data.head()
```

```python
# Column names are viewed:

print(list(london_demo_data.columns))
```

```python
# The data is grouped, aggregated and subset to view the number of BAME residents by borough:

borough_ethnicity = london_demo_data.groupby(['Borough'])['BAME'].agg('sum').sort_values(ascending=False
                                                                                                  ).reset_index()

borough_ethnicity
```

```python
# The results are visualised for easier analysis:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(15, 15)})
ax.set_title("Locations of BAME Residents in London Boroughs (Source: ONS)")
ax.set_xlabel("Borough Name")
ax.set_ylabel("Total BAME Residents")
plt.xticks(rotation=90)

sns.barplot(x='Borough', y='BAME', data=borough_ethnicity)

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')

fig.savefig('borough_BAME_barchart.png')
```

```python
# The same is done, this time to view average household income by borough:

borough_average_income = london_demo_data.groupby(['Borough'])['Mean Annual Household Income estimate (£)'].agg('mean').sort_values(ascending=False
                                                                                                  ).reset_index()

borough_average_income
```

```python
# The oncome and ehtnicity data frames are merged to view how income relates to ethnicity and whether there is
# any correlation between the two.
# Newham clearly has the most BAME residents and the lowest average household income:

borough_income_ethnicity = pd.merge(borough_ethnicity, borough_average_income, how='left', on='Borough')
```

```python
borough_income_ethnicity
```

```python
# The results of this are plotted on a bar plot:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(15, 15)})
ax.set_title("Borough by Income and Number of BAME Residents (Source: ONS)")
ax.set_xlabel("Borough Name")
ax.set_ylabel("Mean Income")
plt.xticks(rotation=90)

sns.barplot(x='Borough', y='BAME', data=borough_ethnicity)

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')

fig.savefig('borough_BAME_barchart.png')
```

```python
# This leads to analysis of an additional data set: road accidents data from tfL for a 10 year period:

road_accidents = pd.read_csv('Road Accidents.csv')

road_accidents.head()
```

```python
# A pivot table in Excel is used to identify the top boroughs for road accidents.
# The data is then subset to include only those boroughs (with an exprected drop during the COVID pandemic):

top_accident_boroughs = ['Westminster', 'Lambeth', 'Tower Hamlets', 'Barnet', 'Southwark', 'Wandsworth', 'Ealing'
                         , 'Croydon', 'Brent', 'Hackney']

top10_accidents = road_accidents[road_accidents['Local authority'].isin(top_accident_boroughs)]

top10_accidents
```

```python
# The results of this are plotted by year on a line plot to see if there is any change by borough by year.
# The plot suggests that there has been a stead increase generall across boroughs over the years:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(15, 15)})
ax.set_title("Top 10 Worst Boroughs for Reported Road Accidents 2010 - 2021 (Source: data.gov.uk)")
ax.set_xlabel("Year")
ax.set_ylabel("Total Accidents")
plt.xticks(rotation=90)

sns.lineplot(x='Accident year', y='All accidents', hue='Local authority', data=top10_accidents)

fig.savefig('top_10_accidents_lineplot.png')
```

```python
# The same is done for all boroughs combined to check this hypothesis:

fig, ax = plt.subplots(dpi=150)

sns.set(rc = {'figure.figsize':(15, 15)})
ax.set_title("Reported London Road Accidents 2010 - 2021 (Source: data.gov.uk)")
ax.set_xlabel("Year")
ax.set_ylabel("Total Accidents")
plt.xticks(rotation=90)

sns.lineplot(x='Accident year', y='All accidents', data=road_accidents, ci=None)

fig.savefig('all_road_accidents_lineplot.png')
```
